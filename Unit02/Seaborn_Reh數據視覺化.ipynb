```python
#導入需要的依賴包
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
```

```python
data = pd.read_csv('D:/ML/IMLP404/ENT/ENT_VT.csv')
```

```python
sns.set(context = 'notebook',style="darkgrid")
plt.xticks(np.arange(0,43,3.5))
a = sns.histplot(data, x="Extrusion_time", binwidth = 3.5, binrange = (0,42))
plt.xlabel('Extrusion Time (in months)')
plt.ylabel('Count')
```

```python
sns.set(context = 'notebook',style="darkgrid")
plt.xticks(np.arange(0,43,3.5))
sns.histplot(data, x="Extrusion_time", binwidth = 3.5, binrange = (0,42), hue = "Ag")
plt.xlabel('Extrusion Time (in months)')
plt.ylabel('Count')
```

```python
sns.set(context = 'notebook',style="darkgrid")
plt.xticks(np.arange(0,43,3.5))
sns.histplot(data, x="Extrusion_time", binwidth = 3.5, binrange = (0,42), hue = "Ag",multiple='stack')
plt.xlabel('Extrusion Time (in months)')
plt.ylabel('Count')
```
